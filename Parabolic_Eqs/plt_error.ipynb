```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("mc_error.csv", delimiter=",", skiprows=1)

n = data[:, 0]
error = data[:, 1]

ref = 1 / np.sqrt(n)
```

```python
plt.figure(figsize=(7, 5))
plt.loglog(n, error, label="Monte Carlo error", linewidth=2)
plt.loglog(n, ref, "--", label=r"$1 / \sqrt{n}$", linewidth=2)

plt.xlabel("Number of samples n")
plt.ylabel("error")
plt.legend()
plt.grid(True, which="both", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```
